```python
# Task 3 : Named Entity Recognition (NER)
from transformers import pipeline

# Load the NER pipeline
ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")

# Sample text for NER
text = "Hugging Face was founded in New York by Clement Delangue."

# Perform NER
entities = ner_pipeline(text)

# Display results
for entity in entities:
    print(f"Entity: {entity['word']}, Label: {entity['entity']}")
```

**Observations**:

The model correctly identifies Hugging Face as an organization, New York as a location, and Clement Delangue as a person.

In some cases, the model may split entities into subwords.

**Explanation**:

The BERT-based model is trained on the CoNLL-2003 dataset, making it effective for entity recognition.

Tokenization Effects: Some entity words may be broken into smaller parts, requiring post-processing.


```python
# Task 4 : Fine-Tune a Model on a Custom Dataset

from datasets import load_dataset

# Load IMDb dataset
dataset = load_dataset("imdb")

# Print a sample from the dataset
print(dataset["train"][0])


from transformers import AutoModelForSequenceClassification, Trainer, TrainingArguments
from transformers import AutoTokenizer
import torch

# Load the tokenizer and model
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    num_train_epochs=2,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    logging_dir="./logs",
    push_to_hub=False
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"].shuffle(seed=42).select(range(1000)),  # Use subset for faster training
    eval_dataset=tokenized_datasets["test"].shuffle(seed=42).select(range(500)),  # Use subset for faster testing
)

# Train the model
trainer.train()
```

**Observations**:

The IMDb dataset contains positive and negative movie reviews.

The model learns to classify reviews with good accuracy after training.

**Explanation**:

DistilBERT is a smaller, faster version of BERT with good accuracy for text classification.

Tokenization ensures the text is properly processed for the model.
Training on a subset (1000 train, 500 test) speeds up fine-tuning.

```python
# Bonus Task: Zero-Shot Classification

from transformers import pipeline

# Load zero-shot classification model
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Sample text
text = "I love learning about deep learning."

# Define candidate labels
labels = ["education", "sports", "technology"]

# Perform zero-shot classification
result = classifier(text, candidate_labels=labels)

# Print results
print(result)
```

**Observations**:

The model assigns high confidence to the correct category "education".

It can classify unseen topics without explicit training.

**Explanation**:

Zero-shot learning allows classification without labeled training data.

BART-based models use a large-scale language model trained on natural inference tasks.
